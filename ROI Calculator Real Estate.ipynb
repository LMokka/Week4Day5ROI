```python
class CashonCashROI:
    def __init__(self):
        self.users = []
    def income_input(self):
        global YearlyIncome
        while True:
            try:
                RentalIncome = float(input("Please enter your monthly rental income   "))
                break
            except ValueError:
                print("Please only input numbers  ")
                continue
        
        while True:
            try:
                LaundryIncome = float(input("Please enter your monthly laundry income  "))
                break
            except ValueError:
                print("Please only input numbers  ")
                continue
        
        TotalMonthlyIncome = RentalIncome + LaundryIncome
        YearlyIncome = TotalMonthlyIncome*12
        print(f"Your total monthly income is $ {TotalMonthlyIncome}")
        print(f"Your total yearly income is $ {YearlyIncome}")
        return(YearlyIncome)
     
    def expenses(self):
        global YearlyExpenses
        while True:
            try:
                PropertyTaxes = float(input("Please enter your monthly property taxes  "))
                break
            except ValueError:
                print("Please only input numbers  ")
                continue
        
        while True:
            try:
                PropertyInsurance = float(input("Please enter your monthly property insurance  "))
                break
            except ValueError:
                print("Please only input numbers  ")
                continue
                
        while True:
            try:
                Electricity = float(input("Please enter your monthly electricity bill  "))
                break
            except ValueError:
                print("Please only input numbers  ")
                continue
            
        while True:
            try:
                Gas = float(input("Please enter your monthly gas bill  "))
                break
            except ValueError:
                print("Please only input numbers  ")
                continue
                
        while True:
            try:
                WaterGarbage = float(input("Please enter your monthly water and garbage bill  "))
                break
            except ValueError:
                print("Please only input numbers  ")
                continue
        
        while True:
            try:
                HOA = float(input("Please enter your monthly HOA bill  "))
                break
            except ValueError:
                print("Please only input numbers  ")
                continue
        
        while True:
            try:
                LawnSnow = float(input("Please enter your monthly lawncare and snow shovelling bill  "))
                break
            except ValueError:
                print("Please only input numbers  ")
                continue
        
        while True:
            try:
                Repairs = float(input("Please enter your monthly repair bill  "))
                break
            except ValueError:
                print("Please only input numbers  ")
                continue
        
        while True:
            try:
                CapExpenses = float(input("Please enter your monthly estimated Capital Expenditure  "))
                break
            except ValueError:
                print("Please only input numbers  ")
                continue
        
        while True:
            try:
                Propertymanagement = float(input("Please enter your monthly estimated Property Management Expenses  "))
                break
            except ValueError:
                print("Please only input numbers  ")
                continue
        
        while True:
            try:
                Mortgage = float(input("Please enter your monthly Mortgage  "))
                break
            except ValueError:
                print("Please only input numbers  ")
                continue
        
        while True:
            try:
                Moneyforvacancy = float(input("Please enter your monthly vacancy expense  "))
                break
            except ValueError:
                print("Please only input numbers  ")
                continue
        
        TotalMonthlyExpenses = PropertyTaxes + PropertyInsurance + Electricity + Gas + WaterGarbage + HOA + LawnSnow + Repairs + CapExpenses + Propertymanagement + Mortgage + Moneyforvacancy
        YearlyExpenses = TotalMonthlyExpenses*12
        print(f"Your total monthly expenses are $ {TotalMonthlyExpenses:.2f}")
        print(f"Your yearly expenses are $ {YearlyExpenses:.2f}")
        return(YearlyExpenses)
        
    def InitialInvestments(self):
        global InitialInvestments
        while True:
            try:
                Downpayment = float(input("Please enter your Downpayment  "))
                break
            except ValueError:
                print("Please only input numbers  ")
                continue
                
        while True:
            try:
                ClosingCosts = float(input("Please enter your Closing Costs  "))
                break
            except ValueError:
                print("Please only input numbers  ")
                continue
        
        while True:
            try:
                RehabBudget = float(input("Please enter your Rehab Budget  "))
                break
            except ValueError:
                print("Please only input numbers  ")
                continue
            
        while True:
            try:
                OtherExpenses = float(input("Please enter your Other Expenses  "))
                break
            except ValueError:
                print("Please only input numbers  ")
                continue
        
        InitialInvestments = Downpayment + ClosingCosts + RehabBudget + OtherExpenses        
        print(f"The Initial Investment for your property is ${InitialInvestments:.2f}")
        return(InitialInvestments)
        
    def ROICalculate(self):
        AnnualCashFlow = YearlyIncome - YearlyExpenses
        print(f"Your Annual Cashflow is ${AnnualCashFlow:.2f}")
        ROI = AnnualCashFlow/InitialInvestments 
        ROI = ROI*100 
        print(f"The ROI for your property is {ROI:0.2f}%")
        return ROI

ROI = CashonCashROI()
ROI.income_input()
ROI.expenses()
ROI.InitialInvestments()
ROI.ROICalculate()
```
